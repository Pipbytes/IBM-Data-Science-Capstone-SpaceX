## Mark NASA John Space Center

We first need to create a folium `Map` object, with an initial center location to be NASA Johnson Space Center at Houston, Texas.


```python
# Start location is NASA Johnson Space Center
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=15)
```

We could use `folium.Circle` to add a highlighted circle area with a text label on a specific coordinate. For example,


```python
# Create an ORANGE circle at NASA Johnson Space Center's coordinate with a popup label showing its name
circle = folium.Circle(nasa_coordinate,
                       radius=1000,
                       color='#d35400',
                       fill=True,
                       fill_opacity=0.2).add_child(folium.Popup('NASA Johnson Space Center'))

# Create a marker at NASA Johnson Space Center's coordinate with a icon showing its name
marker = folium.map.Marker(
    location=nasa_coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 12; color:#d35400;"><b>NASA JSC</b></div>'
        ))

# Add ciricle and marker to the map
site_map.add_child(circle)
site_map.add_child(marker)

site_map
```

and you should find a small yellow circle near the city of Houston and you can zoom-in to see a larger circle.

Recall that data frame spacex_df has detailed launch records, and the `class` column indicates if this launch was successful or not


```python
spacex_df.tail(10)
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


## MarkerCluster

Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

_TODO:_ Create a new column in `launch_sites` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
marker_color = ['green' if i==1 else 'red' for i in spacex_df['class']]

spacex_df['marker_color'] = marker_color
spacex_df.tail(10)
```

```python
# Function to assign color to launch outcome
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'

spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

```python
from os import popen
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

for index, record in spacex_df.iterrows():
    marker = folium.map.Marker(
        location=(record['Lat'],record['Long']),
        popup=folium.Popup(f"Launch site: {record['Launch Site']}<br>Outcome: {'Success' if record['class']==1 else 'Failure'}"),
        icon=folium.Icon(color=record['marker_color'])

    )
    marker_cluster.add_child(marker)

site_map
```

# TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


## MousePosition

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-right) in order to the distance to the launch site.


## Distance function

You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

_TODO:_ Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
ccafs_slc40 = spacex_df[spacex_df['Launch Site'] == 'CCAFS SLC-40']
avg_lat = ccafs_slc40['Lat'].mean()
avg_lon = ccafs_slc40['Long'].mean()
print(avg_lat, avg_lon)
```

```python
# find coordinate of the closet coastline
coastline_lat, coastline_lon = 28.56367, -80.56579

distance_coastline = calculate_distance(avg_lat, avg_lon, coastline_lat, coastline_lon)
distance_coastline
```

## Folium.marker

_TODO:_ After obtained its coordinate, create a `folium.Marker` to show the distance


```python
# add marker for closest coastline point with distance
distance_marker = folium.Marker(
    location=(coastline_lat, coastline_lon),
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance_coastline:10.2f} KM</b></div>',
        # html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_coastline)
    )
)
site_map.add_child(distance_marker)
```

## PolyLine

_TODO:_ Draw a `PolyLine` between a launch site to the selected coastline point


```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
coordinates = [[avg_lat, avg_lon],[coastline_lat, coastline_lon]]
lines = folium.PolyLine(locations=coordinates,
                      weight=2.5,
                      color='red',
                      opacity=0.5)
site_map.add_child(lines)
```

```python
# Coordinates
rail_lat, rail_lon = 28.57203, -80.58527  # Florida East Coast Railway (Cocoa)
highway_lat, highway_lon = 28.5632, -80.57079  # State Road A1A
city_lat, city_lon = 28.3922, -80.6077  # Cape Canaveral


# Calculate distances
distance_rail = calculate_distance(avg_lat, avg_lon, rail_lat, rail_lon)
distance_hwy = calculate_distance(avg_lat, avg_lon, highway_lat, highway_lon)
distance_city = calculate_distance(avg_lat, avg_lon, city_lat, city_lon)
distance_coast = calculate_distance(avg_lat, avg_lon, coastline_lat, coastline_lon)

# Rail way markers
rail_marker = folium.Marker(
    location=(rail_lat, rail_lon),
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance_rail:10.2f} KM</b></div>',
    )
)
site_map.add_child(rail_marker)

# Rail way polyline
rail_lines = folium.PolyLine(
    locations=[[avg_lat, avg_lon], [rail_lat, rail_lon]],
    weight=2.5,
    color='red',
    opacity=0.5
)
site_map.add_child(rail_lines)

# Highway markers
hwy_marker = folium.Marker(
    location=(highway_lat, highway_lon),
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance_hwy:10.2f} KM</b></div>',
    )
)
site_map.add_child(hwy_marker)

# Highway polyline
hwy_lines = folium.PolyLine(
    locations=[[avg_lat, avg_lon], [highway_lat, highway_lon]],
    weight=2.5,
    color='red',
    opacity=0.5
)
site_map.add_child(hwy_lines)

# City markers
city_marker = folium.Marker(
    location=(city_lat, city_lon),
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance_city:10.2f} KM</b></div>',
    )
)
site_map.add_child(city_marker)

# City polyline
city_lines = folium.PolyLine(
    locations=[[avg_lat, avg_lon], [city_lat, city_lon]],
    weight=2.5,
    color='red',
    opacity=0.5
)
site_map.add_child(city_lines)

# Coastline markers
coast_marker = folium.Marker(
    location=(coastline_lat, coastline_lon),
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12px; color:#d35400;"><b>{distance_coast:10.2f} KM</b></div>',
    )
)
site_map.add_child(coast_marker)

# Coastline polyline
coast_lines = folium.PolyLine(
    locations=[[avg_lat, avg_lon], [coastline_lat, coastline_lon]],
    weight=2.5,
    color='blue',
    opacity=0.5
)
site_map.add_child(coast_lines)
```
